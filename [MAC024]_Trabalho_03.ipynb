#  Q1 - Implementação de um  agente  aspirador de pó (100 pontos)


Implementar o simulador de ambiente de medição de desempenho para o um aspirador de pó.
Este ambiente pode ser descrito da seguinte forma:


>* Percepções: Cada aspirador recebe um vetor de percepção de três elementos em cada turno. O primeiro elemento, um sensor de toque/colisão, deve ser 1 se a máquina esbarrar em algo
e um 0 caso contrário. O segundo vem de um fotossensor sob a máquina, que emite
um 1 se houver sujeira e um 0 caso contrário. O terceiro vem de um sensor infravermelho, que
emite 1 quando o agente está em sua localização inicial e 0 caso contrário.

>* Ações: há cinco ações disponíveis: ir em frente, virar 90 graus à direita, 90 graus  à esquerda,
sugar a sujeira e desligar.

>* Objetivos: O objetivo de cada agente é limpar e voltar para casa. Para ser preciso, o desempenho a medida será de 100 pontos para cada pedaço de sujeira aspirado, menos 1 ponto para cada ação realizada e menos 1000 pontos se não estiver no local inicial quando desligar.

>* Ambiente: O ambiente consiste em uma grade de quadrados. Alguns quadrados contêm obstáculos (paredes e móveis) e outras praças são espaços abertos. Alguns dos quadrados abertos podem conter sujeira. Cada ação de "avançar" se move um quadrado, a menos que haja um obstáculo nesse
quadrado, caso em que o agente permaneça onde está, mas o sensor de toque continua. O “sugar sujeira” sempre limpa a sujeira. O comando “desligar” termina a simulação.

Podemos variar a complexidade do ambiente de algumas formas:

> * Forma da sala: No caso mais simples, a sala é um quadrado $n\times n$, para alguns $n$ fixos. nós podemos
tornam mais difícil mudando para uma sala retangular, em forma de L ou de formato irregular,
ou uma série de salas conectadas por corredores.

>* Móveis: Colocar móveis na sala torna-a mais complexa do que uma sala vazia. para o agente de aspiração, uma peça de mobiliário não pode ser distinguida de uma parede por
percepção; ambos aparecem como 1 no sensor de toque/colisão.

>* Colocação da sujeira: No caso mais simples, a sujeira é distribuída uniformemente pela sala. Mas
é mais realista para a sujeira predominar em certos locais, como ao longo de um forte caminho percorrido para a próxima sala, ou na frente do sofá. 



```python
'''
Política: O robô parte da posição inicial definida pelo usuário e verifica se possui algum quadrado vizinho que esteja sujo. Se sim, ele será movido até esse 
quadrado e fará a limpeza. Em seguida, verificará novamente se possui algum vizinho sujo, se sim, irá repetir o processo. Caso não existam vizinhos sujos naquele
quadrado, o robô irá até um vizinho limpo sorteado através da função random. Essa foi uma forma de evitar que o robô tenha apenas um movimento em uma situação em que não existam
quadrados vizinhos que estejam sujos, ou seja, evitar que o robô fique preso em um loop. Essa estratégia causa uma grande variabilidade de movimentos a cada execução do código, mas
consegue limpar todos os quadrados sujos na maioria das vezes. Dessa forma, como existe uma aleatoriedade nos movimentos quando se trata apenas de quadrados limpos, o código ficou
com essa variação a cada execução. Para observar isso, basta ver a pontuação adquirida pelo robô ao final da execução. Por fim, o usuário pode redefinir a sala e o seu tamanho. 
Lembrando que temos: 
l = limpo
s = sujo
o = obstáculo (pode ser parede, móvel, etc.)
'''

import numpy as np
import random

class Aspirador:
    # Criando atributos do aspirador
    def __init__(self):
      self.pontuacao = 0
      self.sensores = [0,0,0]
      self.direcao = 0
      self.modo = 'on'
      self.posicao = [0,0]
      self.verificador = False

    # Definindo funções para limpar quadrado sujo e mover robô. Além disso, atualiza pontuação e sensores de colisão e sujeira 
    def limpa_frente(self):
      self.direcao = 0 # define a direção que será usada pela função "seguir()"
      self.pontuacao += 100 # atualiza pontuação
      self.sensores[0] = 0 # colisão
      self.sensores[1] = 1 # sujeira
      sala[self.posicao[0]][self.posicao[1]+1] = 'l' # remove sujeira
      self.seguir() # chama a função seguir() para fazer o movimento do robô de acordo com as definições acima

    def limpa_atras(self):
      self.direcao = 3
      self.pontuacao += 100
      self.sensores[0] = 0 # colisão
      self.sensores[1] = 1 # sujeira
      sala[self.posicao[0]][self.posicao[1]-1] = 'l' # remove sujeira
      self.seguir()

    def limpa_baixo(self):
      self.direcao = 1
      self.pontuacao += 100
      self.sensores[0] = 0 # colisão
      self.sensores[1] = 1 # sujeira
      sala[self.posicao[0]+1][self.posicao[1]] = 'l' # remove sujeira
      self.seguir()

    def limpa_cima(self):
      self.direcao = 2
      self.pontuacao += 100
      self.sensores[0] = 0 # colisão
      self.sensores[1] = 1 # sujeira
      sala[self.posicao[0]-1][self.posicao[1]] = 'l' # remove sujeira
      self.seguir()

    # Definindo funções para mover robô. Além disso, atualiza pontuação e sensores de colisão e sujeira
    def segue_frente(self):
      self.direcao = 0 
      self.pontuacao -= 1 
      self.sensores[0] = 0 # colisão
      self.sensores[1] = 0 # sujeira
      self.seguir() 

    def segue_atras(self):
      self.direcao = 3
      self.pontuacao -= 1
      self.sensores[0] = 0 # colisão
      self.sensores[1] = 0 # sujeira
      self.seguir()

    def segue_baixo(self):
      self.direcao = 1
      self.pontuacao -= 1
      self.sensores[0] = 0 # colisão
      self.sensores[1] = 0 # sujeira
      self.seguir()

    def segue_cima(self):
      self.direcao = 2
      self.pontuacao -= 1
      self.sensores[0] = 0 # colisão
      self.sensores[1] = 0 # sujeira
      self.seguir()

    # Função responsável por fazer a varredura da sala. Recebe matriz que represanta a sala, além dos valores do tamanho horizontal e vertical da mesma
    def varre_sala(self, sala, n_horizontal, n_vertical):
        # Define uma variável para guardar a posição inicial definida durante a criação do objeto
        posicao_inicial = [0,0]
        posicao_inicial[0] = self.posicao[0]
        posicao_inicial[1] = self.posicao[1]

        while(self.sensores[2] == 0):
          # Agora temos uma sequência de condições para envolver as diversas possibilidades de posições do robô durante a varredura
          if(self.posicao[0] == 0 and self.posicao[1] == 0):
              # Verificamos se algum quadrado vizinho possui sujeira. Se sim, o robô irá até esse quadrado e fará a limpeza
              if(sala[self.posicao[0]][self.posicao[1]+1] == 's'):
                  self.limpa_frente()
              elif(sala[self.posicao[0]+1][self.posicao[1]] == 's'):
                self.limpa_baixo()
              # Caso não exista algum quadrado vizinho que esteja sujo, então o robô apenas irá até um quadrado vizinho que esteja limpo
              elif(sala[self.posicao[0]][self.posicao[1]+1] == 'l' ):
                self.segue_frente()
              elif(sala[self.posicao[0]+1][self.posicao[1]] == 'l' ):
                self.segue_baixo()
          # Faz uma nova verificação sobre outra posição que o robô poderia estar
          elif(self.posicao[0] == n_vertical-1 and self.posicao[1] == n_horizontal-1):
              if(sala[self.posicao[0]][self.posicao[1]-1] == 's'):
                self.limpa_atras()
              elif(sala[self.posicao[0]-1][self.posicao[1]] == 's'):
                self.limpa_cima()
              elif(sala[self.posicao[0]][self.posicao[1]-1] == 'l' ):
                self.segue_atras()
              elif(sala[self.posicao[0]-1][self.posicao[1]] == 'l' ):
                self.segue_cima()
          # Faz uma nova verificação sobre outra posição que o robô poderia estar
          elif(self.posicao[0] == n_vertical-1 and self.posicao[1] == 0):
              if(sala[self.posicao[0]][self.posicao[1]+1] == 's'):
                self.limpa_frente()
              elif(sala[self.posicao[0]-1][self.posicao[1]] == 's'):
                self.limpa_cima()
              elif(sala[self.posicao[0]][self.posicao[1]+1] == 'l' ):
                self.segue_frente()
              elif(sala[self.posicao[0]-1][self.posicao[1]] == 'l' ):
                self.segue_cima()
          # Faz uma nova verificação sobre outra posição que o robô poderia estar
          elif(self.posicao[0] == 0 and self.posicao[1] == n_horizontal-1):
              if(sala[self.posicao[0]][self.posicao[1]-1] == 's'):
                self.limpa_atras()
              elif(sala[self.posicao[0]+1][self.posicao[1]] == 's'):
                self.limpa_baixo()
              elif(sala[self.posicao[0]][self.posicao[1]-1] == 'l' ):
                self.segue_atras()
              elif(sala[self.posicao[0]+1][self.posicao[1]] == 'l' ):
                self.segue_baixo()
          # Faz uma nova verificação sobre outra posição que o robô poderia estar
          elif(self.posicao[1] == n_horizontal-1):
              if(sala[self.posicao[0]][self.posicao[1]-1] == 's'):
                self.limpa_atras()
              elif(sala[self.posicao[0]+1][self.posicao[1]] == 's'):
                self.limpa_baixo()
              elif(sala[self.posicao[0]-1][self.posicao[1]] == 's'):
                self.limpa_cima()
              # Nesse ponto, temos uma mudança. Caso não existam quadrados vizinhos que estejam sujos, o robô irá sortear um vizinho limpo para seguir
              else:
                  while(self.verificador==False):
                      x = random.randint(0, 2)
                      if(x==0):
                          if(sala[self.posicao[0]][self.posicao[1]-1] == 'l' ):
                            self.segue_atras()
                            self.verificador = True
                      elif(x==1):
                          if(sala[self.posicao[0]-1][self.posicao[1]] == 'l' ):
                            self.segue_cima()
                            self.verificador = True
                      elif(x==2):
                          if(sala[self.posicao[0]+1][self.posicao[1]] == 'l' ):
                            self.segue_baixo()
                            self.verificador = True
                  self.verificador = False
          # Faz uma nova verificação sobre outra posição que o robô poderia estar
          elif(self.posicao[1] == 0):
              if(sala[self.posicao[0]][self.posicao[1]+1] == 's'):
                self.limpa_frente()
              elif(sala[self.posicao[0]+1][self.posicao[1]] == 's'):
                self.limpa_baixo()
              elif(sala[self.posicao[0]-1][self.posicao[1]] == 's'):
                self.limpa_cima()
              # Caso não existam quadrados vizinhos que estejam sujos, o robô irá sortear um vizinho limpo para seguir
              else:
                  while(self.verificador==False):
                      x = random.randint(0, 2)
                      if(x==0):
                          if(sala[self.posicao[0]][self.posicao[1]+1] == 'l'):
                            self.segue_frente()  
                            self.verificador = True
                      elif(x==1):
                          if(sala[self.posicao[0]-1][self.posicao[1]] == 'l'):
                            self.segue_cima()  
                            self.verificador = True
                      elif(x==2):
                          if(sala[self.posicao[0]+1][self.posicao[1]] == 'l'):
                            self.segue_baixo()  
                            self.verificador = True
                  self.verificador = False
          # Faz uma nova verificação sobre outra posição que o robô poderia estar
          elif(self.posicao[0] == 0):
              if(sala[self.posicao[0]][self.posicao[1]+1] == 's'):
                self.limpa_frente()
              elif(sala[self.posicao[0]][self.posicao[1]-1] == 's'):
                self.limpa_atras()
              elif(sala[self.posicao[0]+1][self.posicao[1]] == 's'):
                self.limpa_baixo()
              # Caso não existam quadrados vizinhos que estejam sujos, o robô irá sortear um vizinho limpo para seguir
              else:
                  while(self.verificador==False):
                      x = random.randint(0, 2)
                      if(x==0):
                          if(sala[self.posicao[0]][self.posicao[1]+1] == 'l' ):
                            self.segue_frente()
                            self.verificador = True
                      elif(x==1):
                          if(sala[self.posicao[0]][self.posicao[1]-1] == 'l' ):
                            self.segue_atras()  
                            self.verificador = True
                      elif(x==2):
                          if(sala[self.posicao[0]+1][self.posicao[1]] == 'l' ):
                            self.segue_baixo()  
                            self.verificador = True
                  self.verificador = False  
          # Faz uma nova verificação sobre outra posição que o robô poderia estar
          elif(self.posicao[0] == n_vertical-1):
              if(sala[self.posicao[0]][self.posicao[1]+1] == 's'):
                self.limpa_frente()
              elif(sala[self.posicao[0]][self.posicao[1]-1] == 's'):
                self.limpa_atras()
              elif(sala[self.posicao[0]-1][self.posicao[1]] == 's'):
                self.limpa_cima()
              # Caso não existam quadrados vizinhos que estejam sujos, o robô irá sortear um vizinho limpo para seguir
              else:
                  while(self.verificador==False):
                      x = random.randint(0, 2)
                      if(x==0):
                          if(sala[self.posicao[0]][self.posicao[1]+1] == 'l' ):
                            self.segue_frente()
                            self.verificador = True
                      elif(x==1):
                          if(sala[self.posicao[0]][self.posicao[1]-1] == 'l' ):
                            self.segue_atras()
                            self.verificador = True
                      elif(x==2):
                          if(sala[self.posicao[0]-1][self.posicao[1]] == 'l' ):
                            self.segue_cima()  
                            self.verificador = True
                  self.verificador = False
          # Faz uma nova verificação sobre outra posição que o robô poderia estar
          elif(self.posicao[1] == n_horizontal-1):
              if(sala[self.posicao[0]][self.posicao[1]-1] == 's'):
                self.limpa_atras()
              elif(sala[self.posicao[0]+1][self.posicao[1]] == 's'):
                self.limpa_baixo()
              elif(sala[self.posicao[0]-1][self.posicao[1]] == 's'):
                self.limpa_cima()
              elif(sala[self.posicao[0]][self.posicao[1]-1] == 'l' ):
                self.segue_atras()
              elif(sala[self.posicao[0]+1][self.posicao[1]] == 'l' ):
                self.segue_baixo()
              elif(sala[self.posicao[0]-1][self.posicao[1]] == 'l' ):
                self.segue_cima()
          # Caso não existam quadrados vizinhos que estejam sujos, o robô irá sortear um vizinho limpo para seguir
          else:
              if(sala[self.posicao[0]][self.posicao[1]+1] == 's'):
                self.limpa_frente()
              elif(sala[self.posicao[0]-1][self.posicao[1]] == 's'):
                self.limpa_cima()
              elif(sala[self.posicao[0]+1][self.posicao[1]] == 's'):
                self.limpa_baixo()
              elif(sala[self.posicao[0]][self.posicao[1]-1] == 's'):
                self.limpa_atras()
              else:
                  while(self.verificador == False):
                      x = random.randint(0, 3)
                      if(x==0):
                        if(sala[self.posicao[0]][self.posicao[1]+1] == 'l' ):
                          self.segue_frente()
                          self.verificador = True
                      elif(x==1):
                        if(sala[self.posicao[0]][self.posicao[1]-1] == 'l' ):
                          self.segue_atras()
                          self.verificador = True
                      elif(x==2):
                        if(sala[self.posicao[0]+1][self.posicao[1]] == 'l' ):
                          self.segue_baixo()
                          self.verificador = True
                      elif(x==3):
                        if(sala[self.posicao[0]-1][self.posicao[1]] == 'l' ):
                          self.segue_cima()
                          self.verificador = True
                  self.verificador = False
          # Verifica se o robô voltou para a posição inicial. Se sim, então a varredura termina
          if(self.posicao == posicao_inicial):
            self.sensores[2] = 1 # posição inicial
            self.modo = 'off'
            
    # Função para realizar os movimentos. A direção varia de 0 a 3 e o vetor posição guarda o quadrado em que se encontra o robo x em 0 e y em 1
    def seguir(self):
        if self.direcao == 0:
            self.posicao[1] = self.posicao[1] + 1
        if self.direcao == 1:
            self.posicao[0] = self.posicao[0] + 1
        if self.direcao == 2:
            self.posicao[0] = self.posicao[0] - 1
        if self.direcao == 3:
            self.posicao[1] = self.posicao[1] - 1
    
    # Função para verificar se o robô está na posição inicial
    def verificaPosicao(self):
      if(self.posicao[0] == 0 and self.posicao[1] == 0 and self.modo=='off'):
        return True
      else:
        self.pontuacao -= 1000
        return False
    
    # Função para retornar a pontuação obtida pelo robô durante a varredura
    def retornaPontuacao(self):
      return self.pontuacao 

# Constrói a sala
sala = [['l','l','o','o','l','o'],
        ['l','s','o','s','s','l'],
        ['l','s','o','s','o','o'],
        ['o','s','s','s','l','s'],
        ['l','s','o','l','l','o'],
        ['l','l','o','l','o','o']]

print("Sala antes da limpeza do aspirador:")
print(np.matrix(sala))

# Define objeto do tipo Aspirador
asp1 = Aspirador()
# Usa a função varre_sala passando a matriz que representa a sala e os tamanhos como parâmetros
asp1.varre_sala(sala, 6, 6)

print("Sala após a limpeza do aspirador:")
print(np.matrix(sala))
asp1.verificaPosicao()
print("Pontuação: " + str(asp1.retornaPontuacao()))
```
