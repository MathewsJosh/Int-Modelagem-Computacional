#  Q1 - Volume da interseção de três cilindros (30 pontos)


Use o método de Monte Carlo para determinar o volume da interseção de três cilindros com raios unitários.



image.png

```python
############################# PACOTES E BIBLIOTECAS #############################
import math
import numpy as np
import random as r
import matplotlib.pyplot as plt
import pylab as pl
 
#Inicialização de variáveis importantes para o problema
altura_do_cilindro = 8.0
raio = 1.0                #O raio do cilindro é igual ao raio da esfera
 
#Inicialização de variáveis auxiliares
count = 0
intervalos = 1000000
x, y, z = [], [], []
xcaixa, ycaixa, zcaixa = [], [], []
xc1,yc1,zc1 = [], [], []
xc2,yc2,zc2 = [], [], []
xc3,yc3,zc3 = [], [], []
 
 
##################################### MÉTODO #####################################
# Geração de valores randômicos e delimitação dos pontos de cada curva
for i in range(0, intervalos):
  #Geramos valores aleatórios dentro da caixa que simboliza x,y,z
  xaux = r.uniform(-altura_do_cilindro/2, altura_do_cilindro/2)
  yaux = r.uniform(-altura_do_cilindro/2, altura_do_cilindro/2)
  zaux = r.uniform(-altura_do_cilindro/2, altura_do_cilindro/2)
  xcaixa.append(xaux)
  ycaixa.append(yaux)
  zcaixa.append(zaux)

 # Cálculo das curvas dos cilindros e esfera
  curvaCil1 = xaux**2 + yaux**2
  curvaCil2 = zaux**2 + yaux**2
  curvaCil3 = zaux**2 + xaux**2
  curva = xaux**2 + yaux**2 + zaux**2
 
  # Define os pontos pertencentes ao cilindro 1
  if curvaCil1 <= raio and curvaCil1 >= -raio:
    xc1.append(xaux)
    yc1.append(yaux)
    zc1.append(zaux)
    
 
  # Define os pontos pertencentes ao cilindro 2
  if curvaCil2 <= raio and curvaCil2 >= -raio:
    xc2.append(xaux)
    yc2.append(yaux)
    zc2.append(zaux)
 
  # Define os pontos pertencentes ao cilindro 3
  if curvaCil3 <= raio and curvaCil3 >= -raio:
    xc3.append(xaux)
    yc3.append(yaux)
    zc3.append(zaux)
 
  #Pelo método de monte carlo, vamos escolher apenas os pontos randômicos que estão "dentro" da interseção dos cilindros e salva-los em uma lista
  if curvaCil1 <= raio and curvaCil1 >= -raio and curvaCil2 <= raio and curvaCil2 >= -raio and curvaCil3 <= raio and curvaCil3 >= -raio and curva<=raio and curva >= -raio:
    x.append(xaux)
    y.append(yaux)
    z.append(zaux)
 
 
######## Cálculo do volume da esfera #########
print("Raio máximo encontrado no eixo x: " + str(max(x)))
print("Raio máximo encontrado no eixo y: " + str(max(y)))
print("Raio máximo encontrado no eixo z: " + str(max(z)))
 
volume_esfera_unitaria = 4/3*math.pi* 1 * 1 * 1
volume_esfera_montecarlo = 4/3*math.pi* max(x) * max(y) * max(z)
print("O volume para uma esfera de raio 1 é: " + str(volume_esfera_unitaria))
print("O volume da esfera encontrado pelo método de monte carlo, considerando seus pontos de raio máximo é igual a: " + str(volume_esfera_montecarlo) + "\n")
 
 
##################################### Gráficos #####################################
##### Configurações padrão entre os gráficos
scale = altura_do_cilindro/2
size = (7,5)
 
 
######## Gráfico 1 ########
###Configurações dos gráficos
plt.figure(figsize = size)
ax = plt.axes(projection='3d')
ax.set_xlim3d(-scale, scale)
ax.set_ylim3d(-scale, scale)
ax.set_zlim3d(-scale, scale)
ax.plot(xcaixa, ycaixa, zcaixa,'o') #Plota a "caixa" formada pelos valores aleatórios de x,y,z
 
 
######## Gráfico 2 ########
###Configurações dos gráficos
plt.figure(figsize = size)
ax = plt.axes(projection='3d')
ax.set_xlim3d(-scale, scale)
ax.set_ylim3d(-scale, scale)
ax.set_zlim3d(-scale, scale)
ax.plot(xc1, yc1, zc1, 'go')    # Plota os 3 cilindros + esfera que é formada pela sua interseção
ax.plot(xc2, yc2, zc2, 'bo')
ax.plot(xc3, yc3, zc3, 'ro')
ax.plot(x, y, z, 'o')           # Esfera
 
 
######## Gráfico 3 ########
###Configurações dos gráficos
plt.figure(figsize = size)
ax = plt.axes(projection='3d')
ax.set_xlim3d(-scale, scale)
ax.set_ylim3d(-scale, scale)
ax.set_zlim3d(-scale, scale)
ax.plot3D(x, y, z, 'o')           # Plota o gráfico para a esfera
```

# Q2 - Problema com duas urnas (30 pontos)

Use o método de Monte Carlo para reolver o problema abaixo.

A urna I contém 10 bolas brancas 6 y bolas vermelhas. A urna II contém
8 bolas brancas 12  bolas vermelhas. Uma bola ́é  escolhida ao acaso da urna I e colocada na urna II.  

1.  Em seguida, uma bola  ́e retirada ao acaso da urna II. Qual  ́é a  probabilidade desta bola ser branca?


2. Agora considere o mesmo problema, porém são retiradas duas bolas, uma em cada urna:  uma bola ́e retirada ao acaso da urna I e outra retirada da urna II. Qual  ́é a  probabilidade das duas bolas serem vermelhas?


```python
############################# PACOTES E BIBLIOTECAS #############################
import random as r
import pandas as pan
import matplotlib.pyplot as plt

############################# Funções usadas #############################

#Cria a tabela usando a biblioteca PANDAS
def Tabela(valores, index, coluna):
  tabela = pan.DataFrame(valores,index,coluna)
  display(tabela)
  return

#Reabastece os objetos (Caixas e % de cores retiradas) manipulados nas funções
def abastecimento(vazio, refil):
  for i in refil.keys():
    vazio[str(i)] = refil[str(i)]
  return 

#Funções responsaveis por criar e manipular o grafico dos resultados
def plot_results(x, label, real, H_text):
  i = range(0, passo)
  plt.plot(i, x,'.', label=label)
  plt.xlim(-1,passo)
  plt.ylim(0, 100)
  text = ''
  if(label == 'Vermelho'):
    text = 'blue'
  elif(label == 'Branco'):
    text = 'orange'
  else:
    text = 'green'
  plt.axhline(y=real, color=text, linestyle="--")
  plt.annotate(('$'+label+'$'+' medio esperado') ,fontsize = 12,xy=(passo/2, real), xytext=(passo/4, real + H_text))
  plt.grid(True)
  return
def name_grafico(ylabel,xlabel,title):
  plt.title(title, fontsize = 18)
  plt.ylabel(ylabel, labelpad = 18, fontsize = 15)
  plt.xlabel(xlabel, labelpad = 18, fontsize = 15)
  plt.legend(fontsize = 12, frameon = True)
  plt.show()
  return

#constantes
refinamento = 3000
passo = 50
```

```python
#Usando o tipo Dict para auxiliar na organização do código
caixaI = {'brancas': 10, 'vermelhas':6, 'T':16}
caixaII = {'brancas': 8, 'vermelhas':12, 'T':20}
auxI = {}
auxII = {}

MCLI= []
MCLI_refinado = {'vermelho': [], 'branco': []}
saida = {'V_%': [], 'B_%': []}
auxMCLI_refinado = {}
saida_refinada = {'V_p': 0, 'B_p': 0, 'V_%_Vp': [], 'B_%_Vp': [], 'V_%_Bp': [],'B_%_Bp': []}

#testando a probabilidade da 1 questão
print('Retirada 1 bola da 2ª caixa (com '+str(passo)+' iterações e '+str(refinamento)+' testes cada iteração):')
print()
controle = False
for j in range(0,passo):
  abastecimento(auxMCLI_refinado,MCLI_refinado)
  MCLI.clear()
  for i in range(0,refinamento):
    abastecimento(auxI, caixaI)
    abastecimento(auxII, caixaII)
    randon = r.randrange(1, auxI['T'])
    #assumindo que caso o randon for menor que 7 sao vermelhas caso contrario brancas
    if(randon <= auxI['vermelhas']):
      auxI['vermelhas'] -= 1
      auxII['vermelhas'] += 1
      controle = True
      saida_refinada['V_p'] += 1
    else:
      auxI['brancas'] -= 1
      auxII['brancas'] += 1
      controle = False
      saida_refinada['B_p'] += 1
    auxI['T'] -= 1
    auxII['T'] += 1
    r.seed(r.random())
    randon = r.randrange(1, auxII['T'], 1)
    #Agora retirando da caixa II e armazenando os dados para gerar uma %
    if(randon <= auxII['vermelhas']):
      MCLI.append(1)
      if(controle):
        auxMCLI_refinado['vermelho'].append(1)
      else:
        auxMCLI_refinado['branco'].append(1)
    else:
      MCLI.append(0)
      if(controle):
        auxMCLI_refinado['vermelho'].append(0)
      else:
        auxMCLI_refinado['branco'].append(0)
  #Salvamento das medias parciais 
  saida['V_%'].append(sum(MCLI)/refinamento * 100)
  saida['B_%'].append((refinamento - sum(MCLI))/refinamento * 100)

  saida_refinada['V_%_Vp'].append(sum(auxMCLI_refinado['vermelho'])/len(auxMCLI_refinado['vermelho']) * 100)
  saida_refinada['B_%_Vp'].append((len(auxMCLI_refinado['vermelho']) - sum(auxMCLI_refinado['vermelho']))/len(auxMCLI_refinado['vermelho']) * 100)

  saida_refinada['V_%_Bp'].append((sum(auxMCLI_refinado['branco']))/len(auxMCLI_refinado['branco']) * 100)
  saida_refinada['B_%_Bp'].append((len(auxMCLI_refinado['branco']) - sum(auxMCLI_refinado['branco']))/len(auxMCLI_refinado['branco']) * 100)

#Setando os valores de bolas Brancas e Vermelhas retiradas em %  da  caixa
saida_refinada['B_p'] = saida_refinada['B_p']/(refinamento * passo) * 100
saida_refinada['V_p'] = saida_refinada['V_p']/(refinamento * passo) * 100

saida_V = []
saida_B = []

print('Probabilidade de uma cor ser escolhida na 1ª caixa em %')
#Salvando a saida para a criação da primeira tabela (Cores retiradas da 1ª caixa pelo metodo MCL X valores encontrados numericamente ou 'esperados')
saida_V.append(saida_refinada['V_p'])
vermelhas_caixaI = (caixaI['vermelhas']/caixaI['T']) * 100
saida_V.append(vermelhas_caixaI) # numerico

saida_B.append(saida_refinada['B_p'])
brancas_caixaI = (caixaI['brancas']/caixaI['T']) * 100
saida_B.append(brancas_caixaI) # numerico

Tabela(valores=[saida_V, saida_B], index=['Vermelho', 'Branco'], coluna=['MCL', 'Real'])
saida_V.clear()
saida_B.clear()

print()
print('Probalidade da bola retirada da 2 caixa em %')
#Salvando os valores da cor vermelha em %
saida_V.append(sum(saida_refinada['V_%_Vp'])/passo)
saida_V.append(sum(saida_refinada['V_%_Bp'])/passo)
saida_V.append(sum(saida['V_%'])/passo)
Vermelho_real = ((caixaII['vermelhas']/(caixaII['T']+1) * brancas_caixaI)+((caixaII['vermelhas']+1)/caixaII['T'] * vermelhas_caixaI))
saida_V.append( Vermelho_real ) # numerico (Verificar)

#Salvando os valores da cor branca em %
saida_B.append(sum(saida_refinada['B_%_Vp'])/passo)
saida_B.append(sum(saida_refinada['B_%_Bp'])/passo)
saida_B.append(sum(saida['B_%'])/passo)
Branco_real = ((caixaII['brancas']/(caixaII['T']+1) * vermelhas_caixaI)+((caixaII['brancas']+1)/caixaII['T'] * brancas_caixaI))
saida_B.append( Branco_real ) # numerico (Verificar)

#Gerando a tabala com os resultados e um grafico com as medias parciais das iterações
Tabela(valores=[saida_V,saida_B], index=['Vermelho', 'Branco'], coluna=['Vermelho retirado 1º','Branco retirado 1º','Media', 'Real'])
plot_results(saida['V_%'], 'Vermelho', real = (Vermelho_real), H_text=5)
plot_results(saida['B_%'], 'Branco', real= (Branco_real), H_text=5)
name_grafico('Percentual de cores', 'Iterações', 'Porcentagem de cores tiradas em '+ str(refinamento)+' testes X Iterações')

saida_V.clear()
saida_B.clear()
```

```python
#Para o segundo problema usaremos tbm a estrutura de DICT
caixaI = {'brancas': 10, 'vermelhas':6, 'T':16}
caixaII = {'brancas': 8, 'vermelhas':12, 'T':20}
auxI = {}
auxII = {}

MCLII = {'B': 0, 'V': 0, 'VB': 0}
auxMCLII = {}
saida_MCLII = {'B_%': [], 'V_%': [] , 'VB_%': []}

MCLII_refinado = {'V_p':[], 'B_p': [], 'T_V':0, 'T_B':0}
auxMCLII_refinado = {}
saida_MCLII_refinado = {'V_p':{'V':[], 'B':[], 'VB':[]}, 'B_p': {'V':[], 'B':[], 'VB':[]}, 'T_V': [], 'T_B': []}

#testando a probabilidade da 2 questão
print('Retirada 1 bola da 1ª caixa e retirando 1 bola da 2ª caixa (com '+str(passo)+' iterações e '+str(refinamento)+' testes cada iteração):')
print()
controle = False
for j in range(0,passo):
  abastecimento(auxMCLII, MCLII)
  abastecimento(auxMCLII_refinado, MCLII_refinado)
  auxMCLII_refinado['B_p'].clear()
  auxMCLII_refinado['V_p'].clear()
  for i in range(0,refinamento):
    abastecimento(auxI, caixaI)
    abastecimento(auxII, caixaII)
    retirada = r.randrange(1, auxI['T'])
    #assumindo que caso o 'retirada' for menor que 7 sao vermelhas caso contrario brancas
    if(retirada <= auxI['vermelhas']):
      auxI['vermelhas'] -= 1
      auxII['vermelhas'] += 1
      controle = True
      auxMCLII_refinado['T_V'] += 1
    else:
      auxI['brancas'] -= 1
      auxII['brancas'] += 1
      controle = False
      auxMCLII_refinado['T_B'] += 1
    auxI['T'] -= 1
    rauxI = r.randrange(1, auxI['T'])
    auxII['T'] += 1
    rauxII = r.randrange(1, auxII['T'])
    #Agora retirando das duas caixas
    if(rauxI <= auxI['vermelhas'] and rauxII <= auxII['vermelhas']):
      auxMCLII['V'] += 1
      if(controle):
        auxMCLII_refinado['V_p'].append(1)
      else:
        auxMCLII_refinado['B_p'].append(1)
    elif(rauxI > auxI['vermelhas'] and rauxII > auxII['vermelhas']):
      auxMCLII['B'] += 1
      if(controle):
        auxMCLII_refinado['V_p'].append(0)
      else:
        auxMCLII_refinado['B_p'].append(0)
    elif(rauxI <= auxI['vermelhas'] and rauxII > auxII['vermelhas']):
      auxMCLII['VB'] += 1
      if(controle):
        auxMCLII_refinado['V_p'].append(3)
      else:
        auxMCLII_refinado['B_p'].append(3)
    elif(rauxI > auxI['vermelhas'] and rauxII <= auxII['vermelhas']):
      auxMCLII['VB'] += 1
      if(controle):
        auxMCLII_refinado['V_p'].append(3)
      else:
        auxMCLII_refinado['B_p'].append(3)
    else:
      auxMCLII['VB'] += 1
      if(controle):
        auxMCLII_refinado['V_p'].append(3)
      else:
        auxMCLII_refinado['B_p'].append(3)
  #Salvamento das medias parciais
  saida_MCLII['V_%'].append(auxMCLII['V']/refinamento * 100)  
  saida_MCLII['B_%'].append(auxMCLII['B']/refinamento * 100)  
  saida_MCLII['VB_%'].append(auxMCLII['VB']/refinamento * 100)

  saida_MCLII_refinado['V_p']['V'].append(auxMCLII_refinado['V_p'].count(1)/len(auxMCLII_refinado['V_p']) * 100)  
  saida_MCLII_refinado['V_p']['B'].append(auxMCLII_refinado['V_p'].count(0)/len(auxMCLII_refinado['V_p']) * 100)  
  saida_MCLII_refinado['V_p']['VB'].append(auxMCLII_refinado['V_p'].count(3)/len(auxMCLII_refinado['V_p']) * 100) 

  saida_MCLII_refinado['B_p']['V'].append(auxMCLII_refinado['B_p'].count(1)/len(auxMCLII_refinado['B_p']) * 100) 
  saida_MCLII_refinado['B_p']['B'].append(auxMCLII_refinado['B_p'].count(0)/len(auxMCLII_refinado['B_p']) * 100) 
  saida_MCLII_refinado['B_p']['VB'].append(auxMCLII_refinado['B_p'].count(3)/len(auxMCLII_refinado['B_p']) * 100)

  saida_MCLII_refinado['T_V'].append(auxMCLII_refinado['T_V']/refinamento * 100)
  saida_MCLII_refinado['T_B'].append(auxMCLII_refinado['T_B']/refinamento * 100)

saida_V = []
saida_B = []
saida_VB = []

#Valores esperados encontrar pelo metodo MCL
Vermelho_vermelhoCaixaI = ((caixaI['vermelhas']-1)/(caixaI['T']-1) + (caixaII['vermelhas']+1)/(caixaII['T']+1)) /4 *100
Vermelho_brancoCaixaI = ((caixaI['vermelhas'])/(caixaI['T']-1) + (caixaII['vermelhas'])/(caixaII['T']+1)) /4 *100

saida_V.append( Vermelho_vermelhoCaixaI ) #caso tiramos 1 vermelha da primeira caixa (Verificar)
saida_V.append( Vermelho_brancoCaixaI ) #caso tiramos 1 branca da primeira caixa (Verificar)
saida_V.append((saida_V[0] + saida_V[1])/2) #media 

Branco_vermelhaCaixaI = ((caixaI['brancas']-1)/(caixaI['T']-1) + (caixaII['brancas']+1)/(caixaII['T']+1)) /4 *100
Branco_brancoCaixaI = ((caixaI['brancas'])/(caixaI['T']-1) + (caixaII['brancas'])/(caixaII['T']+1)) /4 *100

saida_B.append( Branco_vermelhaCaixaI ) #caso tiramos 1 vermelha da primeira caixa (Verificar)
saida_B.append( Branco_brancoCaixaI ) #caso tiramos 1 branca da primeira caixa (Verificar)
saida_B.append((saida_B[0] + saida_B[1])/2) #media
VermelhoBranco_vermelhoCaixaI = ( (caixaI['brancas'])/(caixaI['T']-1) + (caixaII['vermelhas']+1)/(caixaII['T']+1) +
                                  (caixaI['vermelhas']-1)/(caixaI['T']-1) + (caixaII['brancas'])/(caixaII['T']+1)) /4 *100
VermelhoBranco_brancoCaixaI = ( (caixaI['brancas']-1)/(caixaI['T']-1) + (caixaII['vermelhas'])/(caixaII['T']+1) +
                                  (caixaI['vermelhas'])/(caixaI['T']-1) + (caixaII['brancas']+1)/(caixaII['T']+1)) /4 *100
saida_VB.append( VermelhoBranco_vermelhoCaixaI ) #caso tiramos 1 vermelha da primeira caixa (Verificar)
saida_VB.append( VermelhoBranco_brancoCaixaI ) #caso tiramos 1 branca da primeira caixa (Verificar)
saida_VB.append((saida_VB[0] + saida_VB[1])/2) #media

print('Médias esperadas em %')

#Mostrando uma tabela com os valores numericos que esperamos encontrar
Tabela(valores=[saida_V,saida_B,saida_VB], index=['Vermelhas', 'Brancas','Vermelha&Branca'], coluna=['caso retiremos vermelho da 1ª caixa','caso retiremos branco da 1ª caixa', 'Media'])

saida_V.clear()
saida_B.clear()
saida_VB.clear()
print()
print('Probalidade das bolas retiradas da 1º e 2º caixa em %')

#Salvando os valores se ambas as bolas forem da cor Vermelha
saida_V.append(sum(saida_MCLII_refinado['V_p']['V'])/(passo))
saida_V.append(sum(saida_MCLII_refinado['B_p']['V'])/(passo))
saida_V.append(sum(saida_MCLII['V_%'])/passo)
saida_V.append(sum(saida_MCLII_refinado['T_V'])/passo)#Valor em % da bola vermelha retirada da 1ª caixa

#Salvando os valores se ambas as bolas forem da cor Branca
saida_B.append(sum(saida_MCLII_refinado['V_p']['B'])/(passo))
saida_B.append(sum(saida_MCLII_refinado['B_p']['B'])/(passo))
saida_B.append(sum(saida_MCLII['B_%'])/passo)
saida_B.append(sum(saida_MCLII_refinado['T_B'])/passo) #Valor em % da bola branca retirada da 1ª caixa

#Salvando os valores se uma das bolas for da cor Vermelha e da  cor Branca
saida_VB.append(sum(saida_MCLII_refinado['V_p']['VB'])/(passo))
saida_VB.append(sum(saida_MCLII_refinado['B_p']['VB'])/(passo))
saida_VB.append(sum(saida_MCLII['VB_%'])/passo)
saida_VB.append('-') # Dado o problema esse dado e impossivel 

#Gerando a tabela dos valores obtidos e um grafico com as medias parciais das iterações
Tabela(valores=[saida_V, saida_B, saida_VB], index=['Vermelhas', 'Brancas', 'Vermelha&Branca'], coluna=['Vermelha retirado 1º', 'Branca retirado 1º','Media', '% pego na 1ª caixa no 1º passo'])
plot_results(saida_MCLII['V_%'], 'Vermelho', real= (
    ((((caixaI['vermelhas']-1)/(caixaI['T']-1) + (caixaII['vermelhas']+1)/(caixaII['T']+1)) /2 * Vermelho_vermelhoCaixaI) +
    (((caixaI['vermelhas'])/(caixaI['T']-1) + (caixaII['vermelhas'])/(caixaII['T']+1)) /2 * Vermelho_brancoCaixaI))
), H_text=-10)
plot_results(saida_MCLII['B_%'], 'Branco', real=(
    ((((caixaI['brancas'])/(caixaI['T']-1) + (caixaII['brancas'])/(caixaII['T']+1)) /2 * Branco_vermelhaCaixaI) +
    (((caixaI['brancas']-1)/(caixaI['T']-1) + (caixaII['brancas']+1)/(caixaII['T']+1)) /2 * Branco_brancoCaixaI))
), H_text=5)
plot_results(saida_MCLII['VB_%'], 'Vermelho & Branco', real=(
    ((((caixaI['brancas'])/(caixaI['T']-1) + (caixaII['vermelhas']+1)/(caixaII['T']+1) +
        (caixaI['vermelhas']-1)/(caixaI['T']-1) + (caixaII['brancas'])/(caixaII['T']+1)) /4 * VermelhoBranco_vermelhoCaixaI) + 
    (((caixaI['brancas']-1)/(caixaI['T']-1) + (caixaII['vermelhas'])/(caixaII['T']+1) +
        (caixaI['vermelhas'])/(caixaI['T']-1) + (caixaII['brancas']+1)/(caixaII['T']+1)) /4 * VermelhoBranco_brancoCaixaI))
), H_text=-8)
name_grafico('Percentual de cores', 'Iterações', 'Porcentagem de cores tiradas em '+ str(refinamento)+' testes X Iterações')

saida_V.clear()
saida_B.clear()
saida_VB.clear()
```

# Q3 - Projeto de uma junta adesiva (40 pontos)


O uso de adesivos poliméricos para unir componentes para aplicações estruturais, semiestruturais e não estruturais tem se expandido	 nos últimos anos
como resultado das vantagens que adesivos oferecem para certos processos de construção e reparos.
Existem numerosos tipos de adesivos para várias aplicações e sua classificação depende da natureza química (epoxi, poliuretano, ...), sua forma (líquido, sólido, ...), seu tipo (termosensível, contato, sensível por pressão, ...), e sua capacidade de carga (estrutural, semiestrutural ou não estrutural).

Boas práticas de projeto requerem que as juntas sejam construídas de uma maneira que elas trabalhem sob tensões de cisalhamento ao invés de tensões normais.
A análise mais simples das juntas sugere que a carga aplicada seja uniformemente distribuída sobre a área de ligação. Embora essa análise seja adequada para  juntas rígidas, picos de tensão significativos podem ocorrer em adesivos flexíveis.

shear-lag.png


Um modelo simples de  análise da articulação de juntas flexíveis é  conhecido como modelo {\it shear-lag}. Ele consiste de uma articulação de cola dupla simétrica para ilustrar os princípios e   fornecer informações  sobre as distribuições do tensões de cisalhamento.
A distribuição de tensão de cisalhamento   é dada por
$$ 
\begin{array}{l}
\tau(x) = \frac{P\omega}{4b\sinh(\omega l/2)}\cosh(\omega x) +
\left[  \frac{P\omega}{4b\cosh(\omega l/2)} \left(\frac{2E_ot_o - E_it_i}{2E_ot_o + E_it_i}\right)  
+ \frac{(\alpha_i - \alpha_o)\Delta T \omega}{(1/E_ot_o + 2/E_it_i)\cosh(\omega l/2)}
\right] \sinh(wx)
\end{array}
$$
  onde
  $$
\begin{array}{l}
 \omega = \sqrt{\frac{G}{h} \left(\frac{1}{E_ot_o}+\frac{2}{E_it_i} \right) }
\end{array}
  $$
e $E_o, t_o, \alpha_o$ e $E_i, t_i, \alpha_i$ são o módulo de elasticidade, espessura, coeficiente de expansão térmica do aderente externo ($o$) e interno ($i$), respectivamente; 
$G, h, b$ e $l$ são o módulo de elasticidade transversal, espessura, largura e comprimento do adesivo, respectivamente; 
$\Delta T$ é a mudança de temperatura na junta de ligação
(diferença entre a temperatura de serviço  e a temperatura livre de tensão). Se o adesivo é curado a uma temperatura elevada, de modo que a temperatura livre de tensão da junta seja diferente da temperatura de serviço, a falta de correspondência na expansão térmica dos aderentes externo e interno induz uma tensão térmica através do adesivo).


Considere uma junta dupla mostrada na figura acima que consiste de aderentes externos de alumínio e um aderente interno de aço. A montagem da junta é curada a uma temperatura de 250 $^{\circ}$F e a temperatura livre de  tensões é 200 $^{\circ}$F. A sistema está submetido a uma carga de 2000 lbf e uma temperatura de serviço de 70 $^{\circ}$F. A largura $b$ é igual a 1 in e o comprimento de ligação é de 1 in. Informações adicionais estão tabuladas abaixo:

Para efeito de simplificação, considere uma variabilidade uniforme de 5\% em cada parâmetro do modelo mostrado acima. 

Apresente o seguinte:

- Esboçe o diagrama de tensão de cisalhamento média em função da distância para a região da solda. Para cada ponto e apresente também os valores da média subtraída de 3 desvios-padrões e da média adicionada de três desvios-padrões.

```python
# Importação das bibliotecas necessárias
import numpy as np
import matplotlib.pylab as plt

tensoes_medias = [] # Cria lista para salvar a tensão de cisalhamento média para cada valor de x
xArray = np.arange(0.5,1.5,0.05) # Definição do arranjo para fazer o laço. Vamos usar x variando de 0.5 a 1.5 com um passo de 0.05. Dessa forma, x varia em um intervalo que compreenda o intervalo de ligação (comprimento do adesivo) que é de  1in

for x in xArray:
  # Constantes
  ###################################################
  n = 100 # número de amostras

  # Propriedades Adesivo
  G = 0.2 * 10**6
  G = np.random.uniform(low = G-(G*0.05), high = G+(G*0.05), size=n) 

  alfaad = 55 * 10**(-6)
  alfaad = np.random.uniform(low = alfaad-(alfaad*0.05), high = alfaad+(alfaad*0.05), size=n) 

  h = 0.020
  h = np.random.uniform(low = h-(h*0.05), high = h+(h*0.05), size=n) 

  # Propriedades Aderente Externo
  Eo = 10 *10**(6) #10**7
  Eo = np.random.uniform(low = Eo-(Eo*0.05), high = Eo+(Eo*0.05), size=n) 

  alfao = 13* 10**(-6)
  alfao = np.random.uniform(low = alfao-(alfao*0.05), high = alfao+(alfao*0.05), size=n) 

  to = 0.15
  to = np.random.uniform(low = to-(to*0.05), high = to+(to*0.05), size=n) 

  # Propriedades Aderente Interno
  Ei = 30*10**6
  Ei = np.random.uniform(low = Ei-(Ei*0.05), high = Ei+(Ei*0.05), size=n) 

  alfai = 6* 10**(-6) #13* 6**(-6)
  alfai = np.random.uniform(low = alfai-(alfai*0.05), high = alfai+(alfai*0.05), size=n) 

  ti = 0.1
  ti = np.random.uniform(low = ti-(ti*0.05), high = ti+(ti*0.05), size=n) 

  # Constantes Gerais
  Ts = 70
  Ts = np.random.uniform(low = Ts-(Ts*0.05), high = Ts+(Ts*0.05), size=n)

  Tl = 200
  Tl = np.random.uniform(low = Tl-(Tl*0.05), high = Tl+(Tl*0.05), size=n)

  deltaT = Ts - Tl

  b = 1
  b = np.random.uniform(low = b-(b*0.05), high = b+(b*0.05), size=n)

  l = 1
  l = np.random.uniform(low = l-(l*0.05), high = l+(l*0.05), size=n)

  P = 2000 #P é a carga de 2000 lbf
  P = np.random.uniform(low = P-(P*0.05), high = P+(P*0.05), size=n)

  # Equações
  ###################################################
  w = np.sqrt((G/h)*(1/Eo*to + 2/Ei*ti))
  tau = ((P*w)/(4*b*np.sinh(w*l/2))) * np.cosh(w*x) + (  ((P*w)/(4*b*np.cosh(w*l/2)))*((2*Eo*to - Ei*ti)/(2*Eo*to+Ei*ti)) + (((alfai - alfao)*deltaT*w)/((1/Eo*to + 2/Ei*ti)*np.cosh(w*l/2))))*np.sinh(w*x)
  ###################################################

  # Calcula média das tensões e adiciona na lista tensoes_medias
  tensoes_medias.append(tau.mean())
  # Para cada valor de x, printa o valor de x da iteração e as médias solicitadas no enunciado da questão
  print("Valor de x: " + str(x))
  print("Média subtraída de 3 desvios-padrões: " + str(tau.mean()-(3*tau.std())))
  print("Média adicionada de 3 desvios-padrões: " + str(tau.mean()+(3*tau.std())))
  print("---")

# Plota tensão de cisalhamento média
plt.figure(figsize = (10, 10))
plt.plot(tensoes_medias,xArray)
plt.title("$x$ vs Médias de $τ$", fontsize = 15, pad = 20)
plt.xlabel("$τ$", labelpad = 5, fontsize = 12)
plt.xticks(fontsize = 12)
plt.ylabel("$x$", labelpad = 2, fontsize = 12)
plt.yticks(fontsize = 12)
plt.grid(True)
plt.show()
```
