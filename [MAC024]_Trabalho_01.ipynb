# Evolução de um sistema dinâmico discreto (25 pontos)

Evolua o sistema dinâmico abaixo por 300 passos. Pede-se o seguinte: 

- apresente um gráfico com a evolução das variáveis $x_i(t)$ e $y_i(t)$ ao longo dos passos de tempo simulados e 
-  um gráfico $x\times y$ com os pontos resultantes da simulação realizada.


$$
        \left\{\begin{array}{ccl} x_{i+1} &=&\alpha \left(x_{i} \cos{\left (0.4 - \frac{0.6}{x_{i}^{2} + y_{i}^{2} + 1} \right )} - y_{i} \sin{\left (0.4 - \frac{0.6}{x_{i}^{2} + y_{i}^{2} + 1} \right )}\right) + 1\\y_{i+1} &=&\alpha \left(x_{i} \sin{\left (0.4 - \frac{0.6}{x_{i}^{2} + y_{i}^{2} + 1} \right )} + y_{i} \cos{\left (0.4 - \frac{0.6}{x_{i}^{2} + y_{i}^{2} + 1} \right )}\right)\end{array}\right.,
        \alpha=0.918, (x_0,y_0)=(0,0)
$$

```python
# Importando os pacotes necessários
import numpy as np
import matplotlib.pylab as plt 

# Definimos as variáveis iniciais
x, y = [], []
x.append(0)     #x0=0
y.append(0)     #y0=0
alfa = 0.918
passos = 300

# Laço que faz a evolução do sistema dinâmico
for i in range(0, passos-1):
  func_aux = np.deg2rad(0.4 - (0.6/(x[i]**2 + y[i]**2 + 1)))
  equacao1 = alfa * (x[i] * np.cos(func_aux) - y[i] * np.sin(func_aux)) + 1
  equacao2 = alfa * (x[i] * np.sin(func_aux) + y[i] * np.cos(func_aux))
  x.append(equacao1)
  y.append(equacao2)

# Função responsável por plotar o grafico evolutivo com valores de x e y
def plot_results(x, y, label1, label2, labelX, labelY, title):
    intervalo = range(0, passos)

    plt.figure(figsize = (8, 6))
    plt.plot(x,'-', label=label1)
    plt.plot(y,'-', label=label2)
    plt.xlim(np.amin(intervalo), np.amax(intervalo))
    plt.ylim(np.amin(x) -2, np.amax(x) + 2)

    # Personalização do gráfico (label, fontes, titulos e subtitulos)
    plt.title(title, fontsize = 15, pad = 20)
    plt.xlabel(labelX, labelpad = 5, fontsize = 12)
    plt.xticks(fontsize = 12)
    plt.ylabel(labelY, labelpad = 2, fontsize = 12)
    plt.yticks(fontsize = 12)
    plt.legend(loc = "best", fontsize = 12, frameon = True)
    plt.grid(True)

    plt.show()

# Função responsável por plotar o grafico x*y
def plot_results2(x, y, label1, label2, labelX, labelY, title):

    plt.figure(figsize = (8, 6))
    #plt.plot(x, y,'-', label=label2)
    plt.plot(y, x,'-', label=label1)
    plt.xlim(min(y), max(y))
    plt.ylim(min(x), max(x))

    # Personalização do gráfico (label, fontes, titulos e subtitulos)
    plt.title(title, fontsize = 15, pad = 20)
    plt.xlabel(labelX, labelpad = 5, fontsize = 12)
    plt.xticks(fontsize = 12)
    plt.ylabel(labelY, labelpad = 2, fontsize = 12)
    plt.yticks(fontsize = 12)
    plt.legend(loc = "best", fontsize = 12, frameon = True)
    plt.grid(True)

    plt.show()

# Plotamos os resultados
plot_results(x, y, "x", "y", "Passos", "Evolução", "Gráfico da evolução de x e y em 300 passos")
plot_results2(x, y, "X", "Y", "Y", "X", "Gráfico X × Y")
```

#  Modelo Presa-Predador (25 pontos)


Em 1868, a introdução acidental nos Estados Unidos do inseto de algodão (Icerya purchasi) da Austrália ameaçou destruir a indústria cítrica americana. Para remediar esta situação, um predador australiano natural,  besouro-joaninha (Novius cardinalis), foi importado daquele país. Os besouros mantiveram os insetos em um nível relativamente baixo. Quando o DDT (um inseticida) foi descoberto para matar insetos, os agricultores aplicaram-no na esperança de reduzir ainda mais a população de insetos de algodão. No entanto, o DDT também se mostrou fatal para o besouro, e o efeito geral do uso do inseticida foi aumentar o número dos insetos de algodão. 

Seja  $C_n$ e $B_n$ os níveis de população de insetos de algodão e besouros-joaninha, respectivamente, após $n$ dias. O modelo pode ser escrito da forma
$$
\left\{
  \begin{matrix}
  C_{n+1} &=& C_{n}+  k_{0}C_{n}  -  k_{1}B_{n} C_{n}    \\
  B_{n+1} &=& B_{n}-  k_{2}B_{n}  + k_{3}B_{n} C_{n}    
  \end{matrix}
  \right.
$$
onde $k_i$ são constantes positivas.

- Determine os pontos fixos do sistema  predador-presa. Considere $$(k_0, k_1, k_2, k_3) = (0.6, 0.004, 0.1, 0.001)$$ como parâmetros do modelo.
- Modifique o modelo para refletir um sistema predador-presa em que os agricultores aplicam (com regularidade) um inseticida que destrói tanto o predador do inseto quanto a presa do inseto a uma taxa proporcional aos números presentes. Escolha adequadamente os novos parâmetros do modelo, se assim for necessário.
- Determine os pontos fixos do sistema  predador-presa em que os agricultores aplicam inseticida. 


```python
# Importando as bibliotecas
import numpy as np
import matplotlib.pylab as plt
import sympy as sp
sp.init_printing()

# Inicialização das variáveis
k_0, k_1, k_2, k_3 = sp.var('k_0 k_1 k_2 k_3')
B, C = sp.var('B C')
E = sp.var('E')
v = sp.Matrix([C,B])
fixos = {'k_0':0.6, 'k_1':0.004, 'k_2':0.1, 'k_3':0.001 }

# Funções sem a adição de pesticida
F = sp.Matrix([
     [(C + (k_0*C) - (k_1*B*C))],
     [(B - (k_2*B) + (k_3*B*C))],
    ])

print('1ª Parte')
solucao_1 = sp.solve(F - v,v,dict=True) # Determinação dos pontos fixos
for s in solucao_1:
    display(s)
aux = sp.solve(F.subs(fixos) - v.subs(fixos), v, dict=True)
print('\n Pontos Fixos, substituindo o valor de K:')
for s in aux:
    display(s)

# Funções com a adição de pesticida
F_P = sp.Matrix([
     [(C + (k_0*C) - (k_1*B*C))*((100 - E)/100)],
     [(B - (k_2*B) + (k_3*B*C))*((100 - E)/100)],
    ])

print('\n2ª Parte')
solucao_2 = sp.solve(F_P - v,v,dict=True) # Determinação dos pontos fixos
for s in solucao_2:
    display(s)
print('\nObs: O valor de E na função tem a seguinte condição para que seja visto como porcentagem. 0 <= E <= 100 & E ∈ R')
aux2 = []
valores_E = []
ranger = 41 # Valor sempre menor que #41#, tambem é o numero de vezes que sera executado o programa com uma progressão da efetividade do pesticida
for i in range(ranger): # Atribuindo valor progressivo de E ate 40%
  valores_E.append({'E': (i/(ranger-1) * 40)})
for i in range(ranger): # Calculo substituindo os valores 'K's e 'E'
  aux2.append(sp.solve(F_P.subs(fixos).subs(valores_E[i]) - v.subs(fixos).subs(valores_E[i]), v, dict=True))
print('\nPontos Fixos, substituindo os valores de K e E:')
for i in range(ranger):
    print(str(aux2[i]) + '  Valor de ' + str(valores_E[i]) + '% de efetividade')
```

# Pontos fixos de um sistema dinâmico discreto (25 pontos)

Considere o  sistema dinâmicoa discreto abaixo.
$$\left\{\begin{matrix} x_{i+1} &=&- \delta x_{i} y_{i} + r_{1} x_{i} + x_{i}\\ y_{i+1} &=&- r_{2} y_{i} + \frac{\theta x_{i} y_{i}^{2}}{B + y_{i}} + x_{i}\end{matrix}\right.
$$
onde 
$r_1= 0.5, r_2= 0.1, \delta= 0.01, \theta= 0.01, B= 4, (x_0,y_0)=(20,16).$


-  Apresente os pontos fixos e justifique as suas respectivas classificações quando a estabilidade
- Evolua  o sitema por 1200 iterações e apresente os resultados em um  gráfico  $x \times y$.



<font color='red'> 
A determinação do ponto fixo ocorre quando $x_{i+i} = x_i$


A sugestão é subistituir essa informação na equação acima e usar ou computação numérica ou simbólica para resolver o problema.
</font>

Para resolver o sistema dinâmico e encontrar os pontos fixos irei adotar $x_{i+1} = x_i = x_{eq}$ e $y_{i+1} = y_i = y_{eq}$. Dessa forma, teremos o sistema de equações:

$$\left\{\begin{matrix} x_{eq} &=&- \delta x_{eq} y_{eq} + r_{1} x_{eq} + x_{eq}\\ y_{eq} &=&- r_{2} y_{eq} + \frac{\theta x_{eq} y_{eq}^{2}}{B + y_{eq}} + x_{eq}\end{matrix}\right. \\
$$

Simplificando o sistema para que possa ser utilizada no código:

$$\left\{\begin{matrix} x_{eq} y_{eq} + r_{1} x_{eq}  &=& 0\\ r_{2} y_{eq} + \frac{\theta x_{eq} y_{eq}^{2}}{B + y_{eq}} + x_{eq} - y_{eq}   &=& 0 \end{matrix}\right. \\
$$
Com o sistema simplificado, podemos resolvê-lo através do SymPy e obter os pontos fixos, que são $(0,0)$ e $ (41.0126582278481, 50.0000000000000)$.

```python
# Importação das bibliotecas necessárias
import numpy as np
import matplotlib.pylab as plt  
from sympy import *
import sympy as sym

# Número de iterações
n = 1200
intervalo = np.arange(0,n)

# Constantes
r1 = 0.5
r2 = 0.2
delta = 0.01
theta = 0.01
B = 4
x0 = 20
y0 = 16

# Resolução do Sistema Para encontrar o Ponto Fixo
x,y = sym.symbols('x_eq,y_eq')
eq1 = sym.Eq(-delta*x*y + r1*x,0)
eq2 = sym.Eq(-r2*y +(theta*x*y**2)/(B+y)+ x - y,0)
print("Sistema de equações:\n")
display(eq1)
display(eq2)
print("\nPontos Fixos:")

pontosFixos = sym.solve([eq1,eq2],(x,y))
display(pontosFixos)
print("\n")

# Análise de Estabilidade
f = -delta*x*y +r1*x + x
g = -r2*y +(theta*x*y**2)/(B+y)+ x

# Derivadas das funções do sistema
fx = diff(f,x)
fy = diff(f,y)
gx = diff(g,x)
gy = diff(g,y)
# display(fx)
# display(fy)
# display(gx)
# display(gy)

# Matriz Jacobiana
J = Matrix([[fx,fy],[gx,gy]])
# display(J)

# Calculo do traço da Matriz
trJ = sym.trace(J)
# display(trJ)
trJ = abs(trJ)

# Calculo da determinante da Matriz
detJ = sym.det(J)
# display(detJ)

### Calculo do traço e determinante para cada ponto
# Para o primeiro ponto.
print("Análise do primeiro ponto P = ({},{}):".format(pontosFixos[0][0],pontosFixos[0][1]))
# Traço:
print("\nTraço: ")
traco = trJ.subs([(x,pontosFixos[0][0]),(y,pontosFixos[0][1])])
display(traco)
# Determinante:
print("\nDeterminante: ")
determinante = detJ.subs([(x,pontosFixos[0][0]),(y,pontosFixos[0][1])])
# determinante = detJ.subs(])
display(determinante)
print("\nDeterminante é menor que 1, porém não é maior que o traço menos 1. Dessa forma, o primeiro ponto fixo é instável.")

# Para o segundo ponto.
print("\nAnálise do segundo ponto P = ({},{}):".format(pontosFixos[1][0],pontosFixos[1][1]))
# Traço:
print("\nTraço: ")
traco = trJ.subs([(x,pontosFixos[1][0]),(y,pontosFixos[1][1])])
display(traco)
# Determinante:
print("\nDeterminante: ")
determinante = detJ.subs([(x,pontosFixos[1][0]),(y,pontosFixos[1][1])])
display(determinante)
print("\nDeterminante é menor que 1 e também é maior que o traço menos um. Sendo assim, o segundo ponto fixo é estável.\n")

# Cria vetores solução
x = np.zeros(n)
y = np.zeros(n)

# Inicializa vetores solução
x[0] = x0
y[0] = y0

for i in range(1,n):
  x[i] = -delta*x[i-1]*y[i-1] + r1*x[i-1] + x[i-1]
  y[i] = -r2*y[i-1] + ((theta*x[i-1])*y[i-1]**2)/(B + y[i-1]) + x[i-1]
 

# Gráfico de X e Y
plt.figure(figsize = (8, 6))
plt.plot(x,y)
plt.title("X vs Y", fontsize = 15, pad = 20)
plt.xlabel("$x$", labelpad = 5, fontsize = 12)
plt.xticks(fontsize = 12)
plt.ylabel("$y$", labelpad = 2, fontsize = 12)
plt.yticks(fontsize = 12)
plt.grid(True)
plt.show()
```


# Propagação de um boato em uma empresa (25 pontos)

Considere a divulgação de um boato através de uma empresa de 1500 funcionários, todos trabalhando num
mesmo edifício. Assuma que a propagação do boato é semelhante à disseminação de
uma doença contagiosa, em que o número de pessoas que escuta o boato
 de cada dia é proporcional ao produto do número que ouviu o
boato anteriormente e o número que não ouviu o boato. Isso é dado por
$$
r_{n+1}  = r_{n} +  kr_{n} (1500-r_{n})
$$
onde $k$ é um parâmetro que depende de quão rápido o boato se espalha e $n$ é o número
de dias. Assuma $k =0.0012$ e suponha que quatro pessoas tenham ouvido inicialmente
o boato. Em quanto tempo todos os 1500 funcionários terão ouvido o boato?


Novamente considere a divulgação de um boato, mas agora assuma uma empresa com 5500 funcionários. Considerando o  modelo apresentado acima, construa um modelo para a empresa com as seguintes taxas de crescimento do boato $k=(0.25, 0.025, 0.0025, 0.00025)$  para determinar o número de pessoas que ouviu o boato após 1 semana.


Com base a simulação do modelo anterior, construa uma proposta de como controlar a taxa de crescimento do boato.

```python
# Desenvolva aqui a sua solução. Use mais células se necessário.

# Importando os pacotes necessários
import numpy as np
import matplotlib.pylab as plt

# Função que faz os cálculos para preencher uma lista de resultados de acordo com o modelo 
def propagacao_boato(n_steps, p, a0, k):
  a = np.zeros(n_steps)
  a[0]=a0
  for n in range(0,n_steps-1):
    # Cálculo do valor que será adicionado na lista de resultados
    a[n+1] = a[n] + k*a[n]*(p-a[n])
  return abs(a) # Retorna a lista com valores absolutos

# Função para plotar o gráfico com uma única label
def plota_grafico(lista, n_steps, p, label):
  plt.figure(figsize = (8, 8))
  plt.plot(lista,'o-', label=label)
  plt.xlim(0, n_steps)
  plt.ylim(0, p+500)
  # Personalização do gráfico (label, fontes, titulos e subtitulos)
  plt.axhline(y=p, color='k', linestyle='--', label=p)
  plt.title("Propagação de um boato em uma empresa de "+str(p)+" pessoas", fontsize = 15, pad = 20)
  plt.xlabel("Dias", labelpad = 5, fontsize = 12)
  plt.xticks(fontsize = 12)
  plt.ylabel("Pessoas que ouviram o boato", labelpad = 2, fontsize = 12)
  plt.yticks(fontsize = 12)
  plt.legend(loc = "best", fontsize = 12, frameon = True)
  plt.grid(True)
  plt.show()

# Empresa com 1500 funcionários

# Parâmetros do modelo
k = 0.0012 # Taxa de crescimento do boato
p = 1500 # Número de funcionários
a0 = 4 # Pessoas que ouviram inicialmente o boato
n_steps = 8 # Número de passos do modelo, ou seja, número de dias que serão levados em conta para gerar a lista de resultados

# Usando função propagacao_boato para gerar a lista com os resultados do modelo
resultado = propagacao_boato(n_steps,p, a0, k)
plota_grafico(resultado, n_steps,p,"k = " + str(k))

# Empresa com 5500 funcionários

# Parâmetros do modelo
valores_k = [0.25, 0.025, 0.0025, 0.00025] # Lista com taxas de crescimento do boato
p = 5500
a0 = 4
n_steps = 8

# Plota gráfico com labels diferentes
plt.figure(figsize = (8, 8))
# Adiciona linha pontilhada em x = 5500 para destacar a quantidade de pessoas na empresa
plt.axhline(y=p, color='k', linestyle='--', label=str(p))
# Laço para gerar os resultados com os diferentes valores de k e plotar em um mesmo gráfico
for k in valores_k:
  aux = propagacao_boato(n_steps=n_steps, p=p, a0=a0, k=k)
  plt.plot(aux,'o-', label="k = " + str(k))
  plt.xlim(0, n_steps)
  plt.ylim(0, 12000)
  # Personalização do gráfico (label, fontes, titulos e subtitulos)
  plt.title("Propagação de um boato em uma empresa de "+str(p)+" pessoas", fontsize = 15, pad = 20)
  plt.xlabel("Dias", labelpad = 5, fontsize = 12)
  plt.xticks(fontsize = 12)
  plt.ylabel("Pessoas que ouviram o boato", labelpad = 2, fontsize = 12)
  plt.yticks(fontsize = 12)
  plt.legend(loc = "best", fontsize = 12, frameon = True)
  plt.grid(True)

# Proposta de como controlar a taxa de crescimento do boato

# Parâmetros do modelo
k = 0.000025
p = 5500
a0 = 4
n_steps = 1000

# Plota gráfico com labels diferentes
plt.figure(figsize = (12, 16))
# Adiciona linha pontilhada em com diferentes valores de x para destacar as quantidades de pessoas na empresa
plt.axhline(y=5500, color='k', linestyle='--', label="5500")
plt.axhline(y=3700, color='g', linestyle='--', label="3700")
plt.axhline(y=1900, color='r', linestyle='--', label="1900")
plt.axhline(y=100, color='b', linestyle='--', label="100")
# Laço para gerar o gráfico de empresas com diferentes quantidade de pessoas
for i in range(4):
  aux = propagacao_boato(n_steps=n_steps, p=p, a0=a0, k=k)
  plt.plot(aux,'o-', label="k = " + str(k) + " e p = " + str(p))
  plt.xlim(0, n_steps)
  plt.ylim(0, 7000)
  # Personalização do gráfico (label, fontes, titulos e subtitulos)
  plt.title("Propagação de um boato em uma empresa com diferentes valores de p", fontsize = 15, pad = 20)
  plt.xlabel("Dias", labelpad = 5, fontsize = 12)
  plt.xticks(fontsize = 12)
  plt.ylabel("Pessoas que ouviram o boato", labelpad = 2, fontsize = 12)
  plt.yticks(fontsize = 12)
  plt.legend(loc = "best", fontsize = 12, frameon = True)
  plt.grid(True)
  # Atualiza o valor de p (quantidade de pessoas) a cada iteração
  p = p - 1800
```
